```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from ipywidgets import interact

%matplotlib inline

Tanvir_Apu = Image.open("D:/Class Lecture and Exercise of Sommer Semester 2020/Computer Vision 1 2020/public/Exercise/resources/Tanvir.jpeg")
Tanvir_Apu = np.array(Tanvir_Apu, dtype=np.uint8) # convert Pil Image into a NumPy array and specify that we want values from 0 - 255 instead of 0.0 - 1.0
```

```python
plt.figure()
_ = plt.imshow(Tanvir_Apu) # _ = means that we don't want to save the return value of imshow in a variable either print it out
```

```python
Tanvir_Apu_gray = Tanvir_Apu.mean(axis=-1)
print(Tanvir_Apu_gray.shape)

print(Tanvir_Apu_gray.shape)
print(plt.figure())
# cmap "gray" defines that we want to use a gray color map to represent the pixel values from 0 to 255 
# try to change it to hot or plasma
_ = plt.imshow(Tanvir_Apu_gray, cmap= 'hot')
plt.figure(figsize=(20,20))
```

```python
Tanvir_Apu_gray.shape
```

```python
@interact(x=(-200, 200, 20))
def run(x=0):        
    Tanvir_Apu_new = Tanvir_Apu_gray + x
    plt.figure()  
    plt.imshow(Tanvir_Apu_new, cmap="gray", vmin=0, vmax=255)    
    plt.show()
```
